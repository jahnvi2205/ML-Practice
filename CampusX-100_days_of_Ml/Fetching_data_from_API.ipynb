```python
import requests
import pandas as pd

url = "https://airbnb19.p.rapidapi.com/api/v1/searchDestination"

querystring = {"query":"Chicago","country":"USA"}

headers = {
	"x-rapidapi-key": "YOUR_KEY",
	"x-rapidapi-host": "airbnb19.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

data= response.json()
```

```python
df = pd.DataFrame(data["data"])[['id','location_name','countryCode','display_name']]
df.head()
```
