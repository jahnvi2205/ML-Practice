Liner_regression from scratch

```python
def liner(x,y):
  x_bar= sum(x)/len(x)
  y_bar= sum(y)/len(y)
  # num = summation of (x-x_bar)(y-y_bar)
  # den= summation of (x-x_bar)^2
  num= 0
  den=0
  for i in range(len(x)):
    num+= ((x[i]- x_bar)*(y[i]- y_bar))
    den+= (x[i]- x_bar)**2

  m= num/den
  c= y_bar - m*(x_bar)
  return m,c
```

```python
x=[9,8,4,3,5,6,5,6,10,8,9,8,10,9,8]
y=[8,9,3,4,5,5,6,6,10,6,9,7,9,10,8]
m,c= liner(x,y)
print(m,c)
```

```python
def total_error(x,y,m,c):
  error= 0
  for i in range(len(y)):
    error+= (((m*x[i])+c)- y[i])**2

  return error
```

```python
total_error(x, y, m, c)
```

Linear_regression using scikit-learn library

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data= pd.read_csv('/content/linear_regression_data.csv')
data.head()
data.describe()
```

```python
# iloc: index based location
# df.iloc[rows, columns]

x= data.iloc[:,0:1]
y= data.iloc[:,-1]
x,y
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size= 0.2, random_state=42)
# random_state=42 is standard
```

```python
lr= LinearRegression()
lr.fit(x_train,y_train)
```

```python
print(lr.coef_, lr.intercept_)
```

```python
y_pred = lr.predict(x_test)
```
